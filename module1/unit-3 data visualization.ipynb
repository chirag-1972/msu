# plot marker ,line style, color

```python
#importing matplot lib 
#!pip install matplotlib
import matplotlib.pyplot as plt
# x axis values
x = [1,2,3]
#y axis values 
y = [2,4,1]
#plotting points
plt.plot(x,y,marker=">", c="c", linestyle= 'dashdot', ms = 15, mec ='b', mfc = 'r' )
#mec ='b'border of marker
#mfc = 'r' color of marker
#naming x-axis
plt.xlabel('x-axis')
#naming y-axis
plt.ylabel('y-axis')
#giving title
plt.title('my first graph')
#function to show the plot
plt.show()
```

```python
import numpy as np
x = np.arange(0,10,0.1)
y = np.sin(x)
#figures in matplotlib
plt.figure(figsize=(10,8))
plt.plot(x,y,marker="*", c='g', linestyle='dashed', ms = '10', mec = 'r')
plt.xlabel('x-axis')
plt.ylabel('sin(x)')
plt.title('sine function')
plt.show()
```

```python
import numpy as np 
#plot 1:
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(1,4,1)# subplot(rows,columns,no.of graph[index])
plt.plot(x,y)
plt.title('sub plot 1')
#plot 2:
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(1,4,2)
plt.plot(x,y)
plt.title('sub plot 2')
#plot 3:
x = np.array([0,1,2,3])
y = np.array([1,20,35,67])
plt.subplot(1,4,3)
plt.plot(x,y)
plt.title('sub plot 3')
#plot 4:
x = np.array([0,1,2,3])
y = np.array([87,56,12,67])
plt.subplot(1,4,4)
plt.plot(x,y)
plt.title('sub plot 4')
```

```python
#bar plot 
x=[1,2,3,4,5]
y=[10,5,8,4,2]
plt.bar(x,y, color='b')
```

```python
#scatter plot 
x=[1,2,3,4,5]
y=[10,5,8,4,2]
plt.scatter(x,y, color='b')
```

```python
#histogram used to find frequency of different no.s
data = np.random.normal(170,10,250)
plt.hist(data)
plt.show()
plt.hist(data, bins = 40)
plt.show()
```

```python
#bar plot used to represent categorical data
#ticks parameter
x=['computer','electrical','mechanical','civil','EC']
y=[10,5,8,4,2]
plt.bar(x,y, color='b')
plt.xlabel("engg department")
plt.ylabel("no. of addmission")
plt.title("msu addmission")
plt.tick_params(axis = 'x' , labelsize = 10, labelrotation= 45,labelcolor = 'b')
plt.tick_params(axis = 'y' , labelsize = 15, labelrotation= -10,labelcolor = 'c')
```

# TEXT ANNOTATION 

```python
#draw plot as figure 
fig, ax=plt.subplots()


x = np.linspace(0,20,1000)
y = np.cos(x)
ax.plot(x,y)
ax.axis('equal')#equals both axes
ax.annotate("local maximum", xy=(2*np.pi,1) ,xytext=(10,4),arrowprops=dict(facecolor="black"))
ax.annotate("local minimum", xy=(5*np.pi,-1) ,xytext=(10,-4), arrowprops=dict(facecolor="black"))
plt.show
```
